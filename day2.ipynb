```python
with open("day2.txt", "r") as file:
    l=file.readlines()[0].split(",")
    print(l)

def is_invalid(n):
    if len(str(n))%2!=0:
        return False
    l=len(str(n))
    if str(n)[:l//2]==str(n)[l//2:]:
        return True
sum=0
for x in l:
    st,end=x.split("-")
    for i in range(int(st),int(end)+1):
        if is_invalid(i):
            sum+=i
print(sum)
```

```python
with open("day2.txt", "r") as file:
    l=file.readlines()[0].split(",")
    print(l)

def is_invalid(n):
    l=len(str(n))
    
    for k in range(1, l // 2 + 1):
        if l % k != 0:
            continue
        repeats = l//k
        block = str(n)[:k]
        if block*repeats == str(n):
            return True
    return False

sum=0
for x in l:
    st,end=x.split("-")
    for i in range(int(st),int(end)+1):
        if is_invalid(i):
            sum+=i
print(sum)
```
