```python
with open("day1.txt") as f:
    lines = f.readlines()

s=50
c=0
for line in lines:
    line = line.strip()
    if line[0]=="R":
        s+=int(line[1:])
        c+=abs(s)//100
        if s<=0:
            c+=1
        s%=100 
    elif line[0]=="L":
        if s==0:
            c-=1 #stupid fcking edge case
        s-=int(line[1:])
        c+=abs(s)//100
        if s<=0:
            c+=1
        s%=100
        
print(c)
```
